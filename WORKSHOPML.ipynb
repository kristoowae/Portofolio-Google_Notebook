```python
import pandas as pd

#ubah file csv menjadi dataframe
df = pd.read_csv('Mall_Customers.csv')

# tampilkan 3 baris pertama
df.head(3)
```

```python
# ubah nama kolom
df = df.rename(columns={'Gender': 'gender', 'Age': 'age', 'Annual Income (k$)': 'annual_income', 'Spending Score (1-100)': 'spending_score'})

# ubah data kategorik menjadi data numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)

# tampilkan data yang sudah di preprocess
df.head(3)
```

```python
from sklearn.cluster import KMeans

# menghilangkan kolom customer id dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)

#membuat list yang berisi inertia
clusters = []
for i in range(1,11):
   km = KMeans(n_clusters=i).fit(X)
   clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)

# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_

# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['annual_income'], y=X['spending_score'], hue=X['Labels'],

palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# menghapus kolom yang tidak berguna
df = df.drop(columns=["Provinsi", "Kab/Kota"])
```

```python
# Model SVM


# memisahkan feature dan target
# Klasifikasi Kemiskinan sebagai target
X = df.drop(columns=["Klasifikasi Kemiskinan"])
y = df["Klasifikasi Kemiskinan"]

# Encode the target variable 'y' using LabelEncoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Splitting data train dan data test
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Testing Model
svc_model = SVC(kernel='rbf')
svc_model.fit(X_train, y_train)

# Evaluasi
y_pred = svc_model.predict(X_test)
print("Akurasi SVM:", accuracy_score(y_test, y_pred))
print("Laporan klasifikasi:\n", classification_report(y_test, y_pred))
```

```python
# Model Logistic Regression


# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Normalisasi fitur
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X_train, y_train)

# Prediksi dan evaluasi Logistic Regression
logreg_y_pred = logreg_model.predict(X_test)

print("Akurasi Logistic Regression:", accuracy_score(y_test, logreg_y_pred))
print("Laporan klasifikasi Logistic Regression:\n", classification_report(y_test, logreg_y_pred))
```

```python
# Model Random Forest Classifier


# Inisialisasi dan latih model Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Prediksi dan evaluasi Random Forest
rf_y_pred = rf_model.predict(X_test)

print("Akurasi Random Forest:", accuracy_score(y_test, rf_y_pred))
print("Laporan klasifikasi Random Forest:\n", classification_report(y_test, rf_y_pred))
```

```python
# Model Decision Tree


dec_tree = DecisionTreeClassifier(random_state=42)
dec_tree.fit(X_train, y_train)
y_pred_dec_tree = dec_tree.predict(X_test)
acc_dec_tree = accuracy_score(y_test, y_pred_dec_tree)

print(f"Akurasi Decision Tree: {acc_dec_tree}")
print("Laporan klasifikasi Decision Tree:")
print(classification_report(y_test, y_pred_dec_tree))
```

## **Memilih ML Use Case**

Dalam kasus ini, Machine Learning digunakan untuk menganalisis faktor-faktor yang berkontribusi pada kepuasan pelanggan dan memprediksi pelanggan mana yang berisiko tidak puas atau bahkan churn (berhenti menggunakan layanan).

**Bagaimana ML Bidang Analytics berperan:**

- Identifikasi Faktor Pendorong Kepuasan/Ketidakpuasan: Dengan menggunakan model klasifikasi seperti SVM, kita dapat mengidentifikasi kombinasi spesifik dari durasi_penggunaan, frekuensi_service, skor_kemudahan_booking, skor_kinerja_mekanik, dan skor_harga_terhadap_layanan yang paling berpengaruh terhadap target kepuasan.
- Prediksi Risiko Churn: Model dapat digunakan untuk memprediksi apakah seorang pelanggan baru atau yang sudah ada akan puas (dan kemungkinan loyal) atau tidak puas (dan berisiko churn) berdasarkan data interaksi mereka. Ini memungkinkan perusahaan untuk proaktif dalam menangani pelanggan yang berpotensi tidak puas sebelum mereka benar-benar churn.
- Segmentasi Pelanggan: Hasil analisis dapat membantu dalam mengidentifikasi segmen pelanggan yang berbeda (misalnya, pelanggan yang sangat puas, pelanggan yang berisiko, dll.), memungkinkan strategi retensi yang lebih bertarget.
Rekomendasi Strategi Peningkatan Layanan: Dengan memahami fitur mana yang paling berkorelasi dengan kepuasan, perusahaan dapat memprioritaskan area peningkatan layanan (misalnya, meningkatkan kemudahan booking atau kualitas mekanik) untuk dampak yang maksimal.

Singkatnya, ML di sini bukan hanya untuk membuat prediksi, tetapi juga untuk memberikan wawasan yang mendalam (analytics) tentang mengapa pelanggan puas atau tidak, yang kemudian dapat digunakan untuk pengambilan keputusan strategis dan peningkatan bisnis.
