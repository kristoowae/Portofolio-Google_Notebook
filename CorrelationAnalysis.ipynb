# Library

```python
import numpy as np #library untuk matematika
import pandas as pd #library untuk dataframe
import seaborn as sns #library untuk visualisasi
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

## Load Data

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

## Membuat Scatterplot

```python
sns.scatterplot(data = dataset,
                x = "weight",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "horsepower")
plt.show()
```

## Membuat Korelasi Pearson

```python
# Membuat Korelasi Pearson
# Select only numeric columns for correlation calculation
korelasi = dataset.select_dtypes(include=np.number).corr()
korelasi
```

```python
sns.heatmap(korelasi, vmax = 1, vmin = -1, center = 0)
plt.show()
```

## Uji Hipotesis

```python
from scipy.stats import pearsonr
```

```python
coef, p_val = pearsonr(dataset["mpg"], dataset["displacement"])
```
