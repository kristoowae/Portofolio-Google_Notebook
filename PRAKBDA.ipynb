```python
import urllib3
from bs4 import BeautifulSoup
import xlwt

# Inisialisasi file Excel
result = xlwt.Workbook()
sheet = result.add_sheet("product info")
sheet.write(0, 1, "Product Name")
sheet.write(0, 2, "Price")

# Target URL
url = "https://www.frankana.de/de/multimedia.html"

# Menggunakan PoolManager dari urllib3
http = urllib3.PoolManager()
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
}
response = http.request('GET', url, headers=headers)

# Parsing HTML
soup = BeautifulSoup(response.data, "html.parser")

# Temukan semua produk
all_products = soup.find_all("div", class_="product details product-item-details")

# Simpan hasil ke Excel
index = 0
for product in all_products:
    name_tag = product.find("strong", class_="product name product-item-name")
    price_tag = product.find("span", class_="price")

    if name_tag and price_tag:
        productname = name_tag.text.strip()
        price = price_tag.text.strip()
        index += 1  # <-- perbaiki di sini juga harus += 1, bukan index + 1
        sheet.write(index, 1, productname)
        sheet.write(index, 2, price)

# Simpan file Excel
result.save("product_list.xls")
print(" Produk berhasil disimpan ke 'product_list.xls'")
```

```python
# Melihat Tahapan Preprocessing Text
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

k = "JKN-KIS menanggung perawatan penyakit Orang Dengan Gangguan Jiwa, agar tidak tercipta Joker lainnya :)"
print('kalimat asli')
print(k)
print()

"""
Tahapan Preprocessing
Case Folding
Filtering
Stemming -> dapat refrensi opsional karena menambah kinerja server
Tokenize
Stopword Removal
"""

case_folding = k.lower()
print("Case Folding: ")
print(case_folding)
print()
fil = case_folding.translate(str.maketrans("", "", string.punctuation))
print("Filtering")
print(fil)
print()
print('Stemming')
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemm = stemmer.stem(fil)
print(stemm)
print()
tokens = word_tokenize(stemm)
print("Tokenize")
print(tokens)

stopword = set(stopwords.words('indonesian'))

removed = []
for t in tokens:
    if t not in stopword:
        removed.append(t)
print()
print('stopwords removal')
print(removed)
```

```python
def stdev(nums) :
  diffs = 0
  avg = sum(nums)/len(nums)
  for n in nums:
    diffs += (n - avg)**(2)
  return (diffs/(len(nums)-1))**(0.5)
```
