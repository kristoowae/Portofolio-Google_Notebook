```python
import urllib3
from bs4 import BeautifulSoup
import xlwt

# Inisialisasi file Excel
result = xlwt.Workbook()
sheet = result.add_sheet("product info")
sheet.write(0, 1, "Product Name")
sheet.write(0, 2, "Price")

# Target URL
url = "https://www.frankana.de/de/multimedia.html"

# Menggunakan PoolManager dari urllib3
http = urllib3.PoolManager()
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
}
response = http.request('GET', url, headers=headers)

# Parsing HTML
soup = BeautifulSoup(response.data, "html.parser")

# Temukan semua produk
all_products = soup.find_all("div", class_="product details product-item-details")

# Simpan hasil ke Excel
index = 0
for product in all_products:
    name_tag = product.find("strong", class_="product name product-item-name")
    price_tag = product.find("span", class_="price")

    if name_tag and price_tag:
        productname = name_tag.text.strip()
        price = price_tag.text.strip()
        index += 1  # <-- perbaiki di sini juga harus += 1, bukan index + 1
        sheet.write(index, 1, productname)
        sheet.write(index, 2, price)

# Simpan file Excel
result.save("product_list.xls")
print(" Produk berhasil disimpan ke 'product_list.xls'")
```

### 1.	Algoritma K-Means

Sebelum kita melakukan pengelompokkan data, wajib ditentukan jumlah cluster terbaik dari jumlah data/itearsi yang kita gunakan. Salah satu metode untuk menentukan nilai optimum jumlah cluster dapat digunakan Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Dari tampilan diatas dapat dikatakan jumlah cluster terbaik adalah 2

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=18, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
# Data Cleaning
# Replace '.' with NaN and convert columns to numeric
for col in ['Weight', 'Height']:
    # Changed 'df' to 'data'
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Impute missing values with the mean
data['Weight'].fillna(data['Weight'].mean(), inplace=True)
data['Height'].fillna(data['Height'].mean(), inplace=True)
```

```python
# I will select the 'VIQ' column for clustering.
X = data[['VIQ']]
```

```python
# Finding the best number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Elbow Method graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# From the plot, the elbow is at k=3. So, I will use 3 clusters.
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
# Visualize the clusters
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 7))
sns.scatterplot(data=data, x='VIQ', y=np.zeros(len(data)), hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.title('Clusters of VIQ')
plt.xlabel('VIQ')
plt.yticks([])
plt.legend(title='Cluster')
plt.show()

print(data[['VIQ', 'Cluster']].sort_values(by='VIQ'))
```

```python
# 1. Muat Dataset
# Pastikan file 'iris.csv' berada di direktori yang sama dengan script ini
# atau sesuaikan path_file_anda jika lokasi filenya berbeda.
try:
    df = pd.read_csv('iris.csv')
    print("Dataset 'iris.csv' berhasil dimuat.")
except FileNotFoundError:
    print("Error: File 'iris.csv' tidak ditemukan. Pastikan file berada di direktori yang benar.")
    print("Tidak dapat melanjutkan klasifikasi tanpa dataset.")
    exit() # Keluar jika file tidak ditemukan

# Tampilkan beberapa baris pertama dan nama kolom untuk inspeksi
print("\n--- 5 Baris Pertama Dataset ---")
print(df.head())
print("\n--- Nama Kolom Dataset ---")
print(df.columns)
```

```python
# 2. Pemisahan Fitur (X) dan Target (y)
# Kolom 'species' adalah variabel target
X = df.drop('Species', axis=1) # Fitur adalah semua kolom kecuali 'species'
y = df['Species']             # Target adalah kolom 'species'

print(f"\nJumlah fitur (X): {X.shape[1]}")
print(f"Jumlah sampel (baris): {X.shape[0]}")
print(f"Jenis kelas target (y): {y.unique()}")
```

```python
# 3. Pembagian Data Latih dan Uji
# Membagi data menjadi 70% untuk pelatihan dan 30% untuk pengujian
# random_state untuk hasil yang dapat direproduksi
# stratify=y memastikan distribusi kelas yang sama di set latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nUkuran data pelatihan (X_train): {X_train.shape}")
print(f"Ukuran data pengujian (X_test): {X_test.shape}")
```

```python
# 4. Inisialisasi dan Pelatihan Model Naive Bayes
# Menggunakan Gaussian Naive Bayes karena fitur-fiturnya kontinu
naive_bayes_model = GaussianNB()

print("\nMelatih model Naive Bayes...")
naive_bayes_model.fit(X_train, y_train)
print("Pelatihan model selesai.")
```

```python
# 5. Prediksi pada Data Uji
print("\nMelakukan prediksi pada data pengujian...")
y_pred = naive_bayes_model.predict(X_test)
print("Prediksi selesai.")
```

```python
# 6. Evaluasi Model
print("\n--- Hasil Evaluasi Model Naive Bayes ---")

# Akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy:.4f}")

# Laporan Klasifikasi (Precision, Recall, F1-score untuk setiap kelas)
class_report = classification_report(y_test, y_pred)
print("\nLaporan Klasifikasi:")
print(class_report)

# Opsional: Tampilkan beberapa prediksi dan nilai aktual
print("\n--- Contoh Prediksi vs. Nilai Aktual ---")
hasil_prediksi = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(hasil_prediksi.head())
```

Buatlah visualisasi dengan seaborn menggunakan dataset datairis.csv untuk :

2.	Distribusi Normal

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Muat dataset
try:
    df = pd.read_csv('iris.csv')
except FileNotFoundError:
    print("Error: Pastikan file 'datairis.csv' berada di direktori yang sama.")
    exit()

df.columns = df.columns.str.replace(' ', '') # Menghilangkan spasi dari nama kolom jika ada
if 'SepalLengthCm' in df.columns and 'SepalWidthCm' in df.columns:
    sepal_length_col = 'SepalLengthCm'
    sepal_width_col = 'SepalWidthCm'
elif 'SepalLength' in df.columns and 'SepalWidth' in df.columns:
    sepal_length_col = 'SepalLength'
    sepal_width_col = 'SepalWidth'
else:
    print("Error: Kolom 'SepalLength' atau 'SepalWidth' tidak ditemukan. Harap periksa nama kolom di 'iris.csv'.")
    exit()

# 1. Scatter Plot dan Grafik Garis (SepalLength dan SepalWidth)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df[sepal_length_col], y=df[sepal_width_col], hue='Species', data=df, s=100, alpha=0.7)
sns.lineplot(x=df[sepal_length_col], y=df[sepal_width_col], data=df, color='gray', linestyle='--', alpha=0.5, errorbar=None)
plt.title('Scatter Plot dan Garis: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Species')
plt.show()

# 2. Distribusi Normal (menggunakan Sepal Length sebagai contoh)
plt.figure(figsize=(12, 6))

# Histogram untuk Sepal Length
sns.histplot(df[sepal_length_col], kde=True, color='skyblue', bins=15, stat='density', label='Histogram & KDE')

# Menyesuaikan kurva distribusi normal
mu, std = df[sepal_length_col].mean(), df[sepal_length_col].std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label=f'Distribusi Normal (={mu:.2f}, ={std:.2f})')

plt.title(f'Distribusi Normal Sepal Length ({sepal_length_col})')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Selanjutnya, dua jenis plot dihasilkan. Sebuah garis putus-putus abu-abu juga ditambahkan untuk menunjukkan tren umum, meskipun tanpa error bar untuk menghindari kebingungan. Plot kedua berfokus pada Distribusi Normal dari panjang sepal (SepalLengthCm). Plot ini menggabungkan histogram dan KDE (Kernel Density Estimate) untuk menunjukkan distribusi data aktual, kemudian menempatkan kurva distribusi normal teoretis yang disesuaikan dengan rata-rata () dan standar deviasi () dari data panjang sepal.
